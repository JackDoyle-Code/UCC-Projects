You will now create a library management system by creating a `Library` class and a `Book` class.

Create a class called `Book` with the following attributes:
- Title,
- Author,
- ISBN-13 (i.e. 13-digit International Standard Book Number, see end of lecture 3 notes),
- Number of copies available in your library.

Include a method (i.e. define a function within the class) called `validate()` that return a `True` for a valid ISBN and a `False` for an invalid ISBN.

Define the class such that when you apply the `print()` function, it prints the title of the book.

```python
# Create the Book class here.
class Book:
    def __init__(self, title, author, isbn, copies):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.copies = copies

    def validate(self):
        single_num = str(self.isbn)
        even_list = (single_num[1::2])
        odd_list = (single_num[0::2])
        even = [eval(i) for i in even_list]
        odd = [eval(i) for i in odd_list]
        isbn13 = (sum(even)*3)+sum(odd)
        Result = isbn13 % 10
        if Result == 0:
            return(True)
        else:
            return(False)

    def statement(self):
        return self.title, self.copies
```
