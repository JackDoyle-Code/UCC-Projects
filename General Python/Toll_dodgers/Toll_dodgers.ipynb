```python
# Author - Jack Doyle
# Date of Creation - 27/11/2023
# The function toll dodger: (1) reads in the textfile, 
# (2) calculates the total amount owed by the car registation, 
# (3) taking into account the first day of the month is free 
# and times between 7am and 7pm are €2.50 instead of €1
# (4) prints the list of toll dodgers that owe €50 or more
# in descending order of amount owed

def list_toll_dodgers(tfile):
    textfile = open(tfile, "r")
    lines = textfile.readlines()
    textfile.close() 
# reads in the file and creates a list containing 
# each line in the file. Then closes the file
#bysplittingawhitespacesanddingeachspliotheirrespectiveistshgfsfdgwepppppppp    
    reg = []
    mth = []
    time = []
    for line in lines:
        spt = line.split()
        reg.append(spt[0])
        mth.append(spt[1])
        time.append(spt[2])
# creates 3 lists representing the car registation, 
# date, and time by splitting each line at white spaces 
# and adding each split to their respective lists.

    dic = {}
    for i in reg:
        dic[i] = 0 
# creates a dictionary with each distinct 
# value of reg as the key and the value as zero 
        
    hour = [int(i.split(":")[0]) for i in time]
# creates a list hour by splitting the time list 
# and taking only the hour value
    cost = [2.5 if 7 <= j < 19 else 1 for j in hour]
# creates a cost list containing either 2.5 or 1 
# depending on if the hour is between 7am and 7pm
    multi = [0 if date.split("/")[0] in ["1", "01"] else 1 for date in mth]
# creates a multi list containing either 0 or 1 depending 
# on whether or not it is the first day of the month
    real_cost = [v * cost[k] for k, v in enumerate(multi)]
# creates a real_cost list. This list acounts for the cost of 
# whether the car drove through the toll on the first day of the month.
# It does this by multipling each value in the cost list 
# by their corresponding value in the multi list.
    for m, n in enumerate(real_cost):
            dic[reg[m]] += n
# adds to value corresponding to the car reg for each time it appears            
    sorted_dic = sorted(dic.items(), key=lambda x:x[1], reverse = True)
    sorted_dic = [output for output in sorted_dic if output[1] >= 50.0]
# sorts the dictionary by the values i descending order
# It then runs through each value in the dictionary, checks 
# if its >= 50.0. If true then it adds it to teh sorted_dic list
    print("There are a total of", len(sorted_dic), 
          "toll dodgers with a total cost greater than or equal to €50")
    print("Registration", "|", "Total Cost(€)")
    for i in sorted_dic:
        print(i[0], "|", i[1])
# prints the total number of toll dodgers >= €50
# and lists them in descending order.

list_toll_dodgers("toll_info.txt")
```
